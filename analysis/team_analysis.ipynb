## Team Analysis - Everton FC 2017/2018

In this file we will analyse Everton FC's season. We will look at metrics like Goals Scored, Goals Conceded, Assists, Goal Conversion Rate, Points, Possession, Set Piece Performance and so on...

```python
import matplotlib.pyplot as plt
import numpy as np
from mplsoccer import Pitch, Sbopen
import pandas as pd
import os
import json
import pathlib
```

```python
path = os.path.join(str(pathlib.Path().resolve().parents[0]), 'src', 'wyscout', "events", 'events_England.json') 
with open(path) as f: 
    data = json.load(f) 
events = pd.DataFrame(data) 

#events = events.loc[events.apply (lambda x: len(x.positions) == 2, axis = 1)]

events.head(20)
```

```python
everton_id = 1623
#sub = events.loc[events["eventName"] == "Substitution"].loc[events["teamId"] == everton_id].iloc[0]["index"]

everton_passes = events[(events["teamId"] == everton_id) & (events["eventName"] == "Pass") & (events["matchId"] == 2499723)]
everton_player_ids = everton_passes["playerId"].unique()
print(everton_player_ids)
```

```python
path = os.path.join(str(pathlib.Path().resolve().parents[0]), 'src', 'wyscout', 'players.json') 
with open(path) as f: 
    data = json.load(f) 
players = pd.DataFrame(data) 

players = players.rename(columns={"wyId":"playerId"})
```

```python
merged_df = pd.merge(players, everton_passes, on="playerId")

event_counts = merged_df.groupby('lastName')['eventName'].count()

event_counts
```
